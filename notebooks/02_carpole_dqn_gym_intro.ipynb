V1版本的CartPole

* 目标：通过控制小车的左右移动，使得杆保持直立尽可能长的时间。CartPole 任务被视为已解决（solved）的标准是：连续 100 个回合中，平均奖励 >= 195。

* 状态向量 的 4 个元素分别是：（留意状态向量 ≠ 终止条件）
    * 小车的位置（范围从 -4.8 到 4.8）
    * 小车的速度（范围为无穷到无穷，即无限制）
    * 杆的角度（范围从 -0.41887903 到 0.41887903弧度，即 -24 度到 24 度）
    * 杆的角速度（范围为无穷到无穷，即无限制）

* 动作2（不带有连续的速度）
    * 动作 0：以固定的力向左推小车。
    * 动作 1：以固定的力向右推小车。

* 奖励 (Reward)：每个时间步获得 +1 的奖励，只要杆没有倒下或小车没有超出边界。
* 终止条件：
    * 杆的角度超过 ±12 度。（实际是 ±0.20943951 弧度）
    * 小车的位置超过 ±2.4 的范围（即超过轨道边界）。
    * 达到最大步数（通常为 500 步）。

* 起始状态：所有的观测值被赋予（-0.05，0.05）内的均匀随机值

## 环境


```
import gymnasium as gym

env = gym.make("CartPole-v1", render_mode="rgb_array")

# option不写时的默认参数：low=-0.05, high=0.05
initial_observation, info = env.reset(seed=123, options={"low": -0.1, "high": 0.1})  
```
(array([ 0.03647037, -0.0892358 , -0.05592803, -0.06312564], dtype=float32), {})

* 返回值说明：包含两个元素的元组（初始状态，类型），以及一个空字典{}。
* info一般为特殊时它可能包含一些关于回合结束原因（是正常结束还是被截断）的有用信息，建议直接保留，不用在意

```
env = gym.make("CartPole-v1")

print(env.observation_space)
print(env.action_space)
```

返回：

* 状态空间：`Box([-4.8 -inf -0.41887903 -inf], [4.8 inf 0.41887903 inf], (4,), float32)`
    * 两个array数组，每个包含四个元素，前者是指下界，后者指上界
    * inf指无限
    1. 状态空间是一个box对象
    2. obs_space.shape: (4,)， 类型是元组
    3. obs_space.shape[0],得到4

* 动作空间：`Discrete(2)`
    * box 表示多维连续空间
    * discrete 表示离散值，描述的是可选动作集合，表明存在两个动作，编号分别为 0 和 1。本身并不包含动作的语义和它们产生的物理效果，比如向左向右。其语义是在step方法内，由物理引擎映射好的
    1. 动作空间是一个discrete对象
    2. act_space.n: 2，类型是整数
